```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("sensor_data_600.txt",delimiter = " ",
                   header=None, names =("date","time","ir","lidar"))

data["lidar"].hist(bins=max(data["lidar"]) - min(data["lidar"]), align="left")
plt.show()
```

```python
data.lidar.plot()
plt.show()
```

```python
data["hour"] = [e//10000 for e in data.time]
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()
d.lidar.get_group(14).hist()
plt.show()
```

```python
each_hour = {i:d.lidar.get_group(i).value_counts().sort_index() for i in range(24)}
freqs = pd.concat(each_hour, axis = 1)#concatで連結
freqs = freqs.fillna(0)#欠損値(NaN)を０で埋める
probs = freqs/len(data) #頻度を確率で

probs#表示
```

```python
import seaborn as sns

sns.heatmap(probs)
plt.show()
```

```python
sns.jointplot(data["hour"],data["lidar"],data,kind="kde")
plt.show()
```

```python
p_t = pd.DataFrame(probs.sum())#各列を合計
p_t.plot()
p_t.transpose()#表を横並びに
```

```python
p_t.sum()
```

```python
p_z = pd.DataFrame(probs.transpose().sum())#転置して各列を合計
p_z.plot()
p_z.transpose()
```

```python
##自分の納得用
probs_time_is_0 = freqs[0]/sum(freqs[0])
print(probs_time_is_0)#t=0だけを取り出したセンサ値の確率(条件付き確率？)。値が同時確率より大きいのがわかる。
probs_time_is_0.sum()#足して１になることを確認。
```

```python
cond_z_t = probs/p_t[0]#列ごとにP(t)で割るとP(z|t)になる。
cond_z_t
```

```python
(cond_z_t[6]).plot.bar(color="blue",alpha=0.5)
(cond_z_t[14]).plot.bar(color="orange",alpha=0.5)
plt.show()
```
